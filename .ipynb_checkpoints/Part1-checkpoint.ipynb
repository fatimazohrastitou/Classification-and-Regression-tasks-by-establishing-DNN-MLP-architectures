<h2>Préparation des données et Conception d’un modèle de réseau de neurones en PyTorch</h2>

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset, random_split

# Séparation des caractéristiques et de la variable cible
X = df[['open', 'low', 'high', 'volume']].values
y = df['close'].values
```

```python
# Normalisation des données
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Conversion en tenseurs PyTorch
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32).view(-1, 1)
```

```python
# Création du DataLoader
dataset = TensorDataset(X, y)
train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size
train_dataset, test_dataset = random_split(dataset, [train_size, test_size])
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64)
```

```python

# Définition du modèle de réseau de neurones
class StockPredictor(nn.Module):
    def __init__(self):
        super(StockPredictor, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(4, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 1)
        )
    
    def forward(self, x):
        return self.layers(x)

model = StockPredictor()
criterion = nn.MSELoss()  # Fonction de perte pour la régression
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

<h2> Recherche d'hyperparamètres avec GridSearch </h2>

```python
# Testez différents taux d'apprentissage et architectures manuellement.
learning_rates = [0.01, 0.001, 0.0001]
num_epochs = [50, 100]
```

```python

# Entraînement et évaluation
def train_model(model, train_loader, test_loader, learning_rate, epochs):
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
    train_losses, test_losses = [], []
    
    for epoch in range(epochs):
        model.train()
        for X_batch, y_batch in train_loader:
            optimizer.zero_grad()
            y_pred = model(X_batch)
            loss = criterion(y_pred, y_batch)
            loss.backward()
            optimizer.step()
        
        # Calcul de la perte pour le training et le test
        train_loss, test_loss = 0.0, 0.0
        with torch.no_grad():
            model.eval()
            for X_batch, y_batch in train_loader:
                train_loss += criterion(model(X_batch), y_batch).item()
            for X_batch, y_batch in test_loader:
                test_loss += criterion(model(X_batch), y_batch).item()
                
        train_losses.append(train_loss / len(train_loader))
        test_losses.append(test_loss / len(test_loader))
        
    return train_losses, test_losses

# Exemple d'entraînement avec plusieurs configurations
for lr in learning_rates:
    for epochs in num_epochs:
        model = StockPredictor()
        train_losses, test_losses = train_model(model, train_loader, test_loader, lr, epochs)
        plt.plot(range(epochs), train_losses, label=f'Train Loss (lr={lr}, epochs={epochs})')
        plt.plot(range(epochs), test_losses, label=f'Test Loss (lr={lr}, epochs={epochs})')

plt.xlabel('Époques')
plt.ylabel('Loss')
plt.title('Courbe de la Loss en fonction des Époques')
plt.legend()
plt.show()
```

<h2>Application de techniques de régularisation</h2>

```python
class StockPredictorWithRegularization(nn.Module):
    def __init__(self):
        super(StockPredictorWithRegularization, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(4, 128),
            nn.ReLU(),
            nn.Dropout(0.2),  # Dropout avec un taux de 20%
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(64, 1)
        )
    
    def forward(self, x):
        return self.layers(x)

model = StockPredictorWithRegularization()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)  # Ajout de la régularisation L2

# Entraînez à nouveau le modèle en utilisant la fonction train_model
train_losses, test_losses = train_model(model, train_loader, test_loader, 0.001, 100)

# Visualisation des pertes
plt.plot(range(100), train_losses, label='Train Loss')
plt.plot(range(100), test_losses, label='Test Loss')
plt.xlabel('Époques')
plt.ylabel('Loss')
plt.title('Courbe de la Loss avec Dropout et L2')
plt.legend()
plt.show()
```
