```python
import torch
import torch.nn as nn

class StockPredictor(nn.Module):
    def __init__(self):
        super(StockPredictor, self).__init__()
        self.fc1 = nn.Linear(5, 128)  # 5 input features, 128 neurons in the hidden layer
        self.fc2 = nn.Linear(128, 64) # Example hidden layer
        self.fc3 = nn.Linear(64, 1)   # Adjust the output layer based on your task

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python

# Entraînement et évaluation
def train_model(model, train_loader, test_loader, learning_rate, epochs):
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
    train_losses, test_losses = [], []
    
    for epoch in range(epochs):
        model.train()
        for X_batch, y_batch in train_loader:
            optimizer.zero_grad()
            y_pred = model(X_batch)
            loss = criterion(y_pred, y_batch)
            loss.backward()
            optimizer.step()
        
        # Calcul de la perte pour le training et le test
        train_loss, test_loss = 0.0, 0.0
        with torch.no_grad():
            model.eval()
            for X_batch, y_batch in train_loader:
                train_loss += criterion(model(X_batch), y_batch).item()
            for X_batch, y_batch in test_loader:
                test_loss += criterion(model(X_batch), y_batch).item()
                
        train_losses.append(train_loss / len(train_loader))
        test_losses.append(test_loss / len(test_loader))
        
    return train_losses, test_losses

# Exemple d'entraînement avec plusieurs configurations
for lr in learning_rates:
    for epochs in num_epochs:
        model = StockPredictor()
        train_losses, test_losses = train_model(model, train_loader, test_loader, lr, epochs)
        plt.plot(range(epochs), train_losses, label=f'Train Loss (lr={lr}, epochs={epochs})')
        plt.plot(range(epochs), test_losses, label=f'Test Loss (lr={lr}, epochs={epochs})')

plt.xlabel('Époques')
plt.ylabel('Loss')
plt.title('Courbe de la Loss en fonction des Époques')
plt.legend()
plt.show()
```

<h2>Application de techniques de régularisation</h2>

```python
class StockPredictorWithRegularization(nn.Module):
    def __init__(self):
        super(StockPredictorWithRegularization, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(5, 128),  # Changed from 4 to 5 to match X_batch shape
            nn.ReLU(),
            nn.Dropout(0.2),  # Dropout with a 20% rate
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(64, 1)  # Output layer
        )
    
    def forward(self, x):
        return self.layers(x)
```

```python
model = StockPredictorWithRegularization()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)  # Adding L2 regularization

# Retrain the model using the train_model function
train_losses, test_losses = train_model(model, train_loader, test_loader, 0.001, 100)

# Plot the losses
plt.plot(range(100), train_losses, label='Train Loss')
plt.plot(range(100), test_losses, label='Test Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Curve with Dropout and L2 Regularization')
plt.legend()
plt.show()
```
